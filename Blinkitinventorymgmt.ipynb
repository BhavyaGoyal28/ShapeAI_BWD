```python
#fixing data types

#if not date time
#suppose qty was string

 ### orders["Quantity"] = orders["Quantity"].astype("int")

#if date time
###
orders["OrderDateTime"] = pd.to_datetime(orders["OrderDateTime"])      #### convert obj to datetime format
```

```python
orders["ExpectedDeliveryTime"] = pd.to_datetime(orders["ExpectedDeliveryTime"])      ### to subtract actual delivery time & order time and get the output in the mins.
orders["ActualDeliveryTime"] = pd.to_datetime(orders["ActualDeliveryTime"])         ## dt stands for datetime which is a module in python that cna handle dates and time
```

```python
orders["Actual_Delivery_inmins"] = (orders["ActualDeliveryTime"]-orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
orders["Expected_Delivery_in_mins"] = (orders["ExpectedDeliveryTime"]-orders["OrderDateTime"]).dt.total_seconds()/60
orders
```
